```python
import numpy as np
import matplotlib.pyplot as plt

  
X = ['int8', 'k=2.35','k=2.5','k=2.6','k=2.75','k=3.1', 'k=3.5']
speedup = [
    1.0,
    2.32696634122243,
    2.43346531278231,
    2.57860774,
    2.74286100350328,
    3.0889103,
    3.22633467673149]

acc = [
    75.92,
    76.114,
    76.088,
    76.074,
    76.026,
    74.346,
    73.568    
]

width = 0.35
linewidth=1
fig, ax1 = plt.subplots(figsize=[7, 4])  
X_axis = np.arange(len(X))
plt.grid(axis="y", ls="-.", zorder=3)
b1 = ax1.bar(X_axis, speedup, label = 'Speedup', color="#0257bf", width=width, linewidth=0, edgecolor='black', zorder=2)
ax1.set_ylim(0.5,3.5)
# ax1.legend()
plt.xticks(X_axis, X)
ax1.set_yticks([0.5, 1,2,  3])
ax1.set_ylabel("Speedup", weight='bold')
# ax1.get_yaxis().set_visible(False)
# ax1.legend()
ax1.tick_params(axis="y",direction="in")
ax2 = ax1.twinx()
ax2.set_ylabel("Top-1 Acc", weight='bold')
# ax2.bar(X_axis + width/2, acc, label = 'Acc', color="silver", width=width, linewidth=linewidth, edgecolor='black')
ax2.plot(acc, color="deepskyblue", marker='o', linestyle="--", linewidth=2, markersize=8, label='Acc')
ax2.tick_params(axis="y",direction="in")

ax2.set_ylim(73, 77)
ax2.set_yticks([73, 74, 75, 76, 77])
# ax2.legend()
fig.legend(loc='upper left', bbox_to_anchor=(0,1), bbox_transform=ax1.transAxes)

ax1.spines['left'].set_visible(False)
ax2.spines['left'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax2.spines['right'].set_visible(False)


# plt.show()
plt.savefig('resnet50_acc_speedup_tradeoff.pdf')
```
